```python
a = 10
b = 25
c = 15

if a > b and a > c:
    print("Greatest is:", a)
elif b > c:
    print("Greatest is:", b)
else:
    print("Greatest is:", c)
```

```python
num = 7

if num % 2 == 0:
    print("Even number")
else:
    print("Odd number")
```

```python
ch = 'A'

if ch.isupper():
    print("Uppercase letter")
else:
    print("Lowercase letter")
```

```python
x = '5'

if x.isdigit():
    print("It is a number")
else:
    print("It is a character")
```
